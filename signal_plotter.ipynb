```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python

input = "../dataset/final_cleaned_transformed_dataset.csv"
dataset = pd.read_csv(input)
print(dataset.shape)
```

```python
print(dataset.columns[:20])
```

```python
I_a = dataset["I_A"]
font_size = 25
legend_font_size = 20

plt.figure(figsize = (10, 4))
plt.plot(I_a, label = "I_a_phase")
# plt.title("I_A signal plot")
plt.xlabel("Sample Point", fontsize = font_size)
plt.ylabel("I_A", fontsize = font_size)
plt.xlim(0,10000)
plt.grid(True)
plt.xticks(fontsize=font_size-2)  # Larger x-axis ticks
plt.yticks(fontsize=font_size-2)  # Larger y-axis ticks
plt.legend(loc='lower right', fontsize=legend_font_size) 
plt.savefig("../results/test.png")
```

```python
I_A_var = dataset["I_A_variance"]

plt.figure(figsize = (10, 4))
plt.plot(I_A_var, label = "I_A_variance")
# plt.title("I_A_variance signal plot")
plt.xlabel("Sample Point", fontsize = font_size)
plt.ylabel("I_A_variance", fontsize = font_size)
plt.xlim(0,10000)
plt.grid(True)
plt.xticks(fontsize=font_size-4)  # Larger x-axis ticks
plt.yticks(fontsize=font_size-4)  # Larger y-axis ticks
plt.legend(loc='lower right', fontsize=legend_font_size) 
plt.show()
```

```python
# Normal Case

fourier = "I_A_fft_coeff"
wavelet = "I_A_wavelet_coeff"
I_A_wavelet_coeff = dataset["I_A_wavelet_coeff_6"]
I_A_fft_coeff = dataset["I_A_fft_coeff_6"]

x, y = 15, 7

plt.figure(figsize = (x, y))

xlim0 = 0
xlim1 = 10000
ylim0 = -3.5
ylim1 = 3.5

# plt.subplot(3,1,1)
plt.plot(I_a, label = "I_a")
# plt.title(f"Normal I_a Signal Plot")
plt.xlabel("Sample Point", fontsize = font_size)
plt.ylabel("Current (Amperes)", fontsize = font_size)
plt.ylim(ylim0,ylim1)
plt.xlim(xlim0,xlim1)
plt.grid(True)
plt.xticks(fontsize=font_size-4)  # Larger x-axis ticks
plt.yticks(fontsize=font_size-4)  # Larger y-axis ticks
plt.legend(loc='lower right', fontsize=legend_font_size)
plt.savefig("../results/current_ia.png")



plt.figure(figsize = (x, y))

# plt.subplot(3,1,2)
plt.plot(I_A_fft_coeff, label = fourier)
# plt.title(f"Normal {fourier} Fourier Coefficients Signal Plot")
plt.xlabel("Sample Point", fontsize = font_size)
plt.ylabel("Current (Amperes)", fontsize = font_size)
plt.ylim(ylim0,ylim1)
plt.xlim(xlim0,xlim1)
plt.grid(True)
plt.xticks(fontsize=font_size-4)  # Larger x-axis ticks
plt.yticks(fontsize=font_size-4)  # Larger y-axis ticks
plt.legend(loc='lower right', fontsize=legend_font_size)
plt.savefig("../results/current_ia_fft_coeff_10.png")

plt.figure(figsize = (x, y))

# plt.subplot(3,1,3)
plt.plot(I_A_wavelet_coeff, label = wavelet)
# plt.title(f"Normal {fourier} Wavelet Coefficients Signal Plot")
plt.xlabel("Sample Point", fontsize = font_size)
plt.ylabel("Current (Amperes)", fontsize = font_size)
plt.ylim(ylim0,ylim1)
plt.xlim(xlim0,xlim1)
plt.grid(True)
plt.xticks(fontsize=font_size-4)  # Larger x-axis ticks
plt.yticks(fontsize=font_size-4)  # Larger y-axis ticks
plt.legend(loc='lower right', fontsize=legend_font_size)
plt.savefig("../results/wavelet_normal.png")
```
