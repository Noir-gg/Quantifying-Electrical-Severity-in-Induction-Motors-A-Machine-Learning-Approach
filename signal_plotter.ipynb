```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python

input = "../dataset/final_cleaned_transformed_dataset.csv"
dataset = pd.read_csv(input)
print(dataset.shape)
```

```python
print(dataset.columns[:20])
```

```python
I_a = dataset["I_A"]

plt.figure(figsize = (10, 4))
plt.plot(I_a, label = "I_a_phase")
plt.title("I_A signal plot")
plt.xlabel("Sample Point")
plt.ylabel("I_A")
plt.xlim(0,10000)
plt.grid(True)
plt.legend()
plt.savefig("../results/test.png")
```

```python
ylim0 = -3.2
ylim1 = 3.2
```

```python
# Normal Case

fourier = "I_A_fft_coeff_6"
wavelet = "I_A_wavelet_coeff_6"
I_A_wavelet_coeff_6 = dataset["I_A_wavelet_coeff_6"]
I_A_fft_coeff_1 = dataset["I_A_fft_coeff_6"]

x, y = 15, 5

plt.figure(figsize = (x, y))

xlim0 = 0
xlim1 = 10000

# plt.subplot(3,1,1)
plt.plot(I_a, label = "I_a")
plt.title(f"Normal I_a Signal Plot")
plt.xlabel("Sample Point")
plt.ylabel("Amplitude")
plt.ylim(ylim0,ylim1)
plt.xlim(xlim0,xlim1)
plt.grid(True)
plt.legend(loc='upper right')
plt.savefig("../results/current_ia.png")

plt.figure(figsize = (x, y))

# plt.subplot(3,1,2)
plt.plot(I_A_fft_coeff_1, label = fourier)
plt.title(f"Normal {fourier} Fourier Coefficients Signal Plot")
plt.xlabel("Sample Point")
plt.ylabel("Amplitude")
plt.ylim(ylim0,ylim1)
plt.xlim(xlim0,xlim1)
plt.grid(True)
plt.legend(loc='upper right')
plt.savefig("../results/current_ia_fft_coeff_10.png")

plt.figure(figsize = (x, y))

# plt.subplot(3,1,3)
plt.plot(I_A_wavelet_coeff_6, label = wavelet)
plt.title(f"Normal {wavelet} Wavelet Coefficients Signal Plot")
plt.xlabel("Sample Point")
plt.ylabel("Amplitude")
plt.ylim(ylim0,ylim1)
plt.xlim(xlim0,xlim1)
plt.grid(True)
plt.legend(loc='upper right')
plt.savefig("../results/wavelet_normal.png")

```

```python
# Faulty Case Case

xlim0 = 111000
xlim1 = 121000

fourier = "I_A_fft_coeff_6"
wavelet = "I_A_wavelet_coeff_6"
I_A_wavelet_coeff_6 = dataset["I_A_wavelet_coeff_6"]
I_A_fft_coeff_6 = dataset["I_A_fft_coeff_6"]

x, y = 15, 5
plt.figure(figsize = (x, y))


# plt.subplot(3,1,1)
plt.plot(I_a, label = "I_a")
plt.title(f"Faulty I_a Signal Plot")
plt.xlabel("Sample Point")
plt.ylabel("Amplitude")
plt.ylim(ylim0,ylim1)
plt.xlim(xlim0,xlim1)
plt.grid(True)
plt.legend(loc='upper right')
plt.savefig("../results/1_fault_current_ia.png")

plt.figure(figsize = (x, y))

# plt.subplot(3,1,2)
plt.plot(I_A_fft_coeff_1, label = fourier)
plt.title(f"Faulty {fourier} Fourier Coefficients Signal Plot")
plt.xlabel("Sample Point")
plt.ylabel("Amplitude")
plt.ylim(ylim0,ylim1)
plt.xlim(xlim0,xlim1)
plt.grid(True)
plt.legend(loc='upper right')
plt.savefig("../results/1_fault_current_ia_fft_coeff_6.png")

plt.figure(figsize = (x, y))

# plt.subplot(3,1,3)
plt.plot(I_A_wavelet_coeff_6, label = wavelet)
plt.title(f"Faulty {wavelet} Wavelet Coefficients Signal Plot")
plt.xlabel("Sample Point")
plt.ylabel("Amplitude")
plt.ylim(ylim0,ylim1)
plt.xlim(xlim0,xlim1)
plt.grid(True)
plt.legend(loc='upper right')
plt.savefig("../results/1_fault_current_ia_wavelet_coeff_6.png")

```
